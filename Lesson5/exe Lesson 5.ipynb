```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_generalistes_all = pd.read_excel("Honoraires_totaux_des_professionnels_de_sante_par_departement_en_2016.xls",\
                   sheet_name='Généralistes et MEP')
```

```python
df_generalistes_all.head()
```

```python
df_generalistes_all.shape
```

```python
df_generalistes_all = df_generalistes_all.dropna()
```

```python
df_generalistes_all.shape
```

```python
df_generalistes_all.columns
```

```python
df_generalistes = df_generalistes_all[df_generalistes_all['Généralistes et compétences MEP']
                                      == '01- Médecine générale']
```

```python
df_generalistes.head()
```

```python
df_generalistes['DEPARTEMENT']
```

```python
df_generalistes.dtypes
```

```python
df_generalistes['DEPARTEMENT'].str.contains('^\d', regex=True)
```

```python
df_generalistes = df_generalistes[df_generalistes['DEPARTEMENT'].str.contains(
    '^\d', regex=True)]
```

```python
ax = plt.figure(figsize=(10, 20))
ax = plt.barh(df_generalistes['DEPARTEMENT'], df_generalistes['EFFECTIFS'])
```

```python
df_dentistes_all = pd.read_excel("Honoraires_totaux_des_professionnels_de_sante_par_departement_en_2016.xls",\
                   sheet_name='Dentistes et ODF')

df_dentistes = df_dentistes_all[df_dentistes_all['Chirurgiens-dentistes et ODF']
                                      == '19- Chirurgiens-dentistes']
df_dentistes = df_dentistes[df_dentistes['DEPARTEMENT'].str.contains(
    '^\d', regex=True)]

plt.figure(figsize=(10, 20))
plt.barh(df_dentistes['DEPARTEMENT'], df_dentistes['EFFECTIFS'])
```

```python
df_sages_all = pd.read_excel("Honoraires_totaux_des_professionnels_de_sante_par_departement_en_2016.xls",\
                   sheet_name='Sages-femmes')

df_sages = df_sages_all[df_sages_all['Sages-femmes']
                                      == '21- Sages-femmes']
df_sages = df_sages[df_sages['DEPARTEMENT'].str.contains(
    '^\d', regex=True)]

plt.figure(figsize=(10, 20))
plt.barh(df_sages['DEPARTEMENT'], df_sages['EFFECTIFS'])
```
